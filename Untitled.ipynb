# Top 10 Countries by COVID-19 lLeast number of Cases per Million

```python
val = res['response']
ans = []
for i in val:
    temp = [] 
    a = i['country']  
    b = i['cases']['1M_pop']
    if b is not None:
        try:
            b_int = int(b.replace(',', ''))  
            temp.append(b_int)
            temp.append(a)
            ans.append(temp)
        except :
            pass

ans.sort()
country = []
TestPerMillion = []
for i in range(10) :
    country.append(ans[i][1])
    TestPerMillion.append(ans[i][0])
```

```python
plt.ylabel('Country')
plt.xlabel('Cases Per Million')
plt.title('Top 10 Countries by COVID-19 Lowest Cases per Million')
plt.gcf().autofmt_xdate() 
plt.barh( country , TestPerMillion, color='green' , alpha=0.25)
plt.figure(figsize=(20,20))
```
